```python
# TASK: Create a 2x2 subplot layout.
# Plot a line chart in the first subplot, a bar chart in the second, a scatter plot in the third, and a pie chart in the fourth.

# TASK: Create a 2x2 subplot layout.
# Plot a line chart in the first subplot, a bar chart in the second, a scatter plot in the third, and a pie chart in the fourth.

import matplotlib.pyplot as plt
import numpy as np

# Data for the plots
x = np.arange(0, 11)
y = x**2
categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 9]
scatter_x = [1, 2, 3, 4, 5]
scatter_y = [2, 4, 6, 8, 10]
pie_labels = ['Python', 'Java', 'C++', 'JavaScript']
pie_sizes = [40, 25, 20, 15]

# Create the subplot layout
fig, axs = plt.subplots(2, 2, figsize=(10, 10))

# Line chart in the first subplot
axs[0, 0].plot(x, y, marker='o')
axs[0, 0].set_title('Line Plot of y = x^2')
axs[0, 0].set_xlabel('x values')
axs[0, 0].set_ylabel('y values')
axs[0, 0].grid(True)

# Bar chart in the second subplot
axs[0, 1].bar(categories, values, color=['red', 'blue', 'green', 'purple'])
axs[0, 1].set_title('Bar Plot of Categories and Values')

# Scatter plot in the third subplot
axs[1, 0].scatter(scatter_x, scatter_y, color='blue')
for i in range(len(scatter_x)):
    axs[1, 0].annotate(f'({scatter_x[i]}, {scatter_y[i]})', (scatter_x[i], scatter_y[i]), textcoords="offset points", xytext=(0,10), ha='center')
axs[1, 0].set_title('Scatter Plot with Annotations')

# Pie chart in the fourth subplot
axs[1, 1].pie(pie_sizes, labels=pie_labels, autopct='%1.1f%%', startangle=140)
axs[1, 1].set_title('Pie Chart of Programming Languages')
axs[1, 1].axis('equal')

# Adjust layout
plt.tight_layout()

# Display the plot
plt.show()
```
